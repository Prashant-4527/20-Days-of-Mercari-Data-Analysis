# Featuring Engineerning

```python
import numpy as np
```

```python
# Extarct main_category and sub_category
df["main_category"] = df["category_name"].str.split("/").str[0]
df["sub_category"] = df["category_name"].str.split("/").str[1]
```

```python
# Map condition to label
condition_map = {1: "New", 2: "Like New", 3: "Good", 4: "Fair", 5:"Poor"}
df["condition_label"] = df["item_condition_id"].map(condition_map)
```

# Numpy On Mercari

--------------------------------------- NUMPY FEATURE ENGINEERING ---------------------------------------------

```python
# Log price transformation
df["log_price"] = np.log1p(df["price"])

print("=== Price vs Log Price ===")
print(f"Raw price  — mean: {df['price'].mean():.2f}, std: {df['price'].std():.2f}, max: {df['price'].max():.2f}")
print(f"Log price  — mean: {df['log_price'].mean():.4f}, std: {df['log_price'].std():.4f}, max: {df['log_price'].max():.4f}")
```

```python
# IQR outlier detection on price
Q1 = np.percentile(df["price"], 25)
Q3 = np.percentile(df["price"], 75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"\n=== IQR Outlier Analysis ===")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Normal range: {lower:.2f} to {upper:.2f}")

outliers = df[(df["price"] < lower) | (df["price"] > upper)]
print(f"Outlier count: {outliers.shape[0]}")
print(f"Outlier %: {outliers.shape[0]/len(df)*100:.2f}%")
print(f"\nTop 5 outlier prices:")
print(outliers.nlargest(5, "price")[["name", "price", "main_category"]])
```

```python
# Flag outliers with np.where
df["is_price_outlier"] = np.where(
    (df["price"] < lower)  |   (df["price"] > upper), 1,0
)
print(f"\nOutlier flag distribution:")
print(df["is_price_outlier"].value_counts())
```

```python
# Price tier with np.where (faster than apply)
df["price_tier"] = np.where(
    df["price"] <= 1000, "budget",
    np.where(df["price"] <=50000, "mid", "premium")
)


print(f"\n=== Price Tier Distribution ===")
print(df["price_tier"].value_counts())
print(df.groupby("price_tier")["price"].mean().round(2))
```

```python
# Percentile analysis per category
print("\n=== 90th Percentile Price per Category ===")
p90 = df.groupby("main_category")["price"].apply(
    lambda x: np.percentile(x, 90)
).round(2).sort_values(ascending=False)
print(p90)
```

```python
# 6. Final feature summary
print(f"\n=== Final Dataset Shape ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print(f"New NumPy features: log_price, is_price_outlier, price_tier")
```

# Advanced Charts On Mercari

```python
# Mercari color palette — assign meaning, not decoration
COLORS = {
    "primary":   "#4F46E5",   # main bars, main lines
    "secondary": "#0EA5E9",   # comparison, secondary data
    "accent":    "#F59E0B",   # highlights, important values
    "positive":  "#10B981",   # good outcomes, growth
    "negative":  "#EF4444",   # problems, drops, warnings
    "neutral":   "#6B7280",   # gridlines, reference lines
}
```

```python
def style_ax(ax, title, xlabel, ylabel, rotate_x=False):
    """Apply consistent Mercari project styling to any axes object."""
    ax.set_title(title, fontsize=13, fontweight="bold", pad=12)
    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel(ylabel, fontsize=11)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(axis="y", alpha=0.3, color="gray", linestyle="--")
    ax.set_axisbelow(True)
    ax.tick_params(labelsize=9)
    if rotate_x:
        ax.tick_params(axis="x", rotation=45)
```

------- CHART 1: Price Distribution - linear vs log --------------------------------------------

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df["price"], bins=80,
             color=COLORS["primary"], edgecolor="white")
style_ax(axes[0], "Price — linear scale",
         "Price ($)", "Number of items")

axes[1].hist(df["log_price"], bins=80,
             color=COLORS["positive"], edgecolor="white")
style_ax(axes[1], "Price — log scale (better)",
         "log(1 + price)", "Number of items")

fig.suptitle("Mercari Price Distribution", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
COLORS = {
    "primary":   "#4F46E5",
    "secondary": "#0EA5E9",
    "accent":    "#F59E0B",
    "positive":  "#10B981",
    "negative":  "#EF4444",
    "neutral":   "#6B7280",
}

def style_ax(ax, title, xlabel, ylabel, rotate_x=False):
    ax.set_title(title, fontsize=13, fontweight="bold", pad=12)
    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel(ylabel, fontsize=11)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(axis="y", alpha=0.3, color="gray", linestyle="--")
    ax.set_axisbelow(True)
    ax.tick_params(labelsize=9)
    if rotate_x:
        ax.tick_params(axis="x", rotation=45)
```

 ── CHART A: Price distribution ───────────────────────────

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df["price"], bins=80,
             color=COLORS["primary"], edgecolor="white")
style_ax(axes[0], "Price — Linear (skewed)",
         "Price ($)", "Count")

axes[1].hist(df["log_price"], bins=80,
             color=COLORS["positive"], edgecolor="white")
style_ax(axes[1], "Log Price — Normal shape",
         "log(1 + Price)", "Count")

fig.suptitle("A — Why we use log price transformation",
             fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

── CHART B: Category — count vs avg price ────────────────

```python
cat_stats = df.groupby("main_category")["price"].agg(
    count="count", avg_price="mean"
).sort_values("avg_price", ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].bar(cat_stats.index, cat_stats["count"],
            color=COLORS["secondary"], edgecolor="white")
style_ax(axes[0], "Listing Volume per Category",
         "Category", "Number of Items", rotate_x=True)

axes[1].bar(cat_stats.index, cat_stats["avg_price"],
            color=COLORS["accent"], edgecolor="white")
overall_avg = df["price"].mean()
axes[1].axhline(overall_avg, color=COLORS["negative"],
                linestyle="--", linewidth=1.5,
                label=f"Overall avg ${overall_avg:.2f}")
axes[1].legend(fontsize=9)
style_ax(axes[1], "Average Price per Category",
         "Category", "Avg Price ($)", rotate_x=True)

fig.suptitle("B — Volume vs Value: which categories matter most?",
             fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

Women has 45% of listings but Electronics has 2x higher avg price → Volume ≠ value. Different categories need different business strategies → Mercari should apply category-specific seller fees and features.

── CHART C: Condition vs price ───────────────────────────

```python
condition_order = ["New", "Like New", "Good", "Fair", "Poor"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x="condition_label", y="log_price",
            order=condition_order, hue="condition_label",
            palette="muted", legend=False, ax=axes[0])
axes[0].set_title("Price Distribution by Condition",
                  fontsize=13, fontweight="bold")
axes[0].set_xlabel("Condition")
axes[0].set_ylabel("log(1 + Price)")

condition_avg = df.groupby("condition_label")["price"].mean()\
                  .reindex(condition_order)
axes[1].bar(condition_avg.index, condition_avg.values,
            color=COLORS["primary"], edgecolor="white")
style_ax(axes[1], "Avg Price by Condition",
         "Condition", "Avg Price ($)")

fig.suptitle("C — Does condition predict price?",
             fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

Price doesn't cleanly decrease from New → Poor due to small sample sizes in Poor condition → Condition alone doesn't predict price reliably → Combine condition with category for stronger signal.

── CHART D: Shipping analysis ────────────────────────────

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

shipping_avg = df.groupby("shipping")["price"].mean()
axes[0].bar(["Buyer pays (0)", "Seller pays (1)"],
            shipping_avg.values,
            color=[COLORS["neutral"], COLORS["positive"]],
            edgecolor="white")
style_ax(axes[0], "Avg Price by Shipping Type",
         "Who Pays Shipping", "Avg Price ($)")

shipping_cat = df.groupby(["main_category","shipping"])["price"]\
                 .mean().unstack().round(2)
shipping_cat.plot(kind="bar", ax=axes[1],
                  color=[COLORS["neutral"], COLORS["positive"]],
                  edgecolor="white", width=0.7)
axes[1].set_title("Avg Price by Category + Shipping",
                  fontsize=13, fontweight="bold")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Avg Price ($)")
axes[1].tick_params(axis="x", rotation=45)
axes[1].legend(["Buyer pays", "Seller pays"], fontsize=9)
axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)

fig.suptitle("D — Shipping behavior reveals seller strategy",
             fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

── CHART E: Correlation heatmap ──────────────────────────

```python
numeric_cols = ["price","log_price","name_length",
                "has_description","item_condition_id","shipping"]
corr = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            vmin=-1, vmax=1, square=True, linewidths=0.5, ax=ax)
ax.set_title("E — Feature Correlation Matrix",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

 ── CHART F: Top brands by avg price ─────────────────────

```python
top_brands = df[df["brand_name"] != "no brand"]\
               .groupby("brand_name")\
               .filter(lambda x: len(x) >= 3)\
               .groupby("brand_name")["price"].mean()\
               .nlargest(10).round(2)

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(top_brands.index, top_brands.values,
               color=COLORS["accent"], edgecolor="white")
for bar, val in zip(bars, top_brands.values):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f"${val:.0f}", va="center", fontsize=9)
ax.set_title("F — Top 10 Brands by Average Price",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Avg Price ($)")
ax.set_ylabel("Brand")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.show()
```
