# Data Cleaning 

```python
print("Shape before cleaning:", df.shape)
```

```python
print("\nNull %:\n", (df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Drop Rows where category_name is null (only 0.43% = safe to drop)
df = df.dropna(subset=["category_name"])
print("\nAfter dropping null categories:", df.shape)
```

```python
# Fill Brand_name nulls with "No Brand" (43% Data - Never Drop)
df["brand_name"] = df["brand_name"].fillna("No brand")
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check price = 0 rows - are these real?
print("\nItems with price = 0:")
df[df["price"] == 0][["name", "category_name", "brand_name"]].head(10)
```

```python
# item description have some null lets drop them
df = df.dropna(subset=["item_description"])
```

```python
# Verify Cleaning worked:
print(f"Null counts after cleaning: {df.isnull().sum()}")
```

# Featuring Engineerning

```python
import numpy as np
```

```python
# Extarct main_category and sub_category
df["main_category"] = df["category_name"].str.split("/").str[0]
df["sub_category"] = df["category_name"].str.split("/").str[1]
```

```python
# Normalize brand_name to lowercase for consistent grouping
df["brand_name"] = df["brand_name"].str.lower().str.strip()
```

```python
# has_description - is the description meaningful?
df["has_description"] = df["item_description"].apply(
    lambda x: 0 if x == "No description yet" else 1
)
```

```python
# name_length - longer name might indicate more detail = higher price?
df["name_length"] = df["name"].str.len()
```

```python
# Map condition to label
condition_map = {1: "New", 2: "Like New", 3: "Good", 4: "Fair", 5:"Poor"}
df["condition_label"] = df["item_condition_id"].map(condition_map)
```
