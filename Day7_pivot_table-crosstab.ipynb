# Data Cleaning 

```python
print("Shape before cleaning:", df.shape)
```

```python
print("\nNull %:\n", (df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Drop Rows where category_name is null (only 0.43% = safe to drop)
df = df.dropna(subset=["category_name"])
print("\nAfter dropping null categories:", df.shape)
```

```python
# Fill Brand_name nulls with "No Brand" (43% Data - Never Drop)
df["brand_name"] = df["brand_name"].fillna("No brand")
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check price = 0 rows - are these real?
print("\nItems with price = 0:")
df[df["price"] == 0][["name", "category_name", "brand_name"]].head(10)
```

```python
# item description have some null lets drop them
df = df.dropna(subset=["item_description"])
```

```python
# Verify Cleaning worked:
print(f"Null counts after cleaning: {df.isnull().sum()}")
```

# Featuring Engineerning

```python
import numpy as np
```

```python
# Extarct main_category and sub_category
df["main_category"] = df["category_name"].str.split("/").str[0]
df["sub_category"] = df["category_name"].str.split("/").str[1]
```

```python
# Normalize brand_name to lowercase for consistent grouping
df["brand_name"] = df["brand_name"].str.lower().str.strip()
```

```python
# has_description - is the description meaningful?
df["has_description"] = df["item_description"].apply(
    lambda x: 0 if x == "No description yet" else 1
)
```

```python
# name_length - longer name might indicate more detail = higher price?
df["name_length"] = df["name"].str.len()
```

```python
# Map condition to label
condition_map = {1: "New", 2: "Like New", 3: "Good", 4: "Fair", 5:"Poor"}
df["condition_label"] = df["item_condition_id"].map(condition_map)
```

# groupby & aggregation

--------GROUPBY ANALYSIS---------------------------------

```python
# Avergae price per main_category - sorted descending
print("===  Avg Price by Category  ===")
df.groupby("main_category")["price"].mean().sort_values(ascending=False).round(2)
```

```python
# Item count per main_category
print("\n=== Item Count by Category ===")
df.groupby("main_category").size().sort_values(ascending=False)
```

```python
# Full agg tabel - avg, count, max per category
print("\n========             Full Category Summary            ===========")
summary = df.groupby("main_category")["price"].agg(
    avg_price="mean",
    total_item="count",
    max_price="max",
    min_price="min"
).round(2).sort_values("avg_price", ascending=False)
summary
```

```python
# Does shipping affect price?
print("\n=== Avg Price by Shipping ===")
df.groupby("shipping")["price"].mean().round(2)
```

```python
# Avg Price by Condition
print("\n=== Avg Price by Condition ===")
df.groupby("condition_label")["price"].mean().round(2).sort_values(ascending=False)
```

```python
# Multi-column: avg price by category + shipping
print("===  Avg Price by Category + Shipping  ===")
df.groupby(["main_category", "shipping"])["price"].mean().sort_values(ascending=False).round(2)
```

# merge & concat

```python
# Create a category metadata table (simulating a second table)
cat_meta = pd.DataFrame({
    "main_category": ["Women", "Beauty", "Kids", "Electronics", 
                      "Men", "Home", "Vintage & Collectibles", 
                      "Other", "Handmade", "Sports & Outdoors"],
    "market_segment": ["Fashion", "Personal Care", "Family", "Tech",
                       "Fashion", "Lifestyle", "Collectibles",
                       "Misc", "Artisan", "Active"],
    "avg_platform_fee": [0.10, 0.12, 0.10, 0.08, 
                         0.10, 0.11, 0.13,
                         0.10, 0.15, 0.11]
})
```

```python
# Left join - attach category metadata to listings 
df_enriched = pd.merge(df, cat_meta, on="main_category", how="left")
```

```python
print("Original shape:", df.shape)
print("Enriched shape:", df_enriched.shape)
print("\nNew columns:", [c for c in df_enriched.columns if c not in df.columns])
print("\nSample:\n", df_enriched[["name", "main_category", "market_segment", "avg_platform_fee"]].head(10))
print("\nUnmatched categories (NaN market_segment):")
print(df_enriched[df_enriched["market_segment"].isnull()]["main_category"].value_counts())
```

# pivot_table & crosstab

---------------------- PIVOT TABLE ANALYSIS -------------------------------------------

```python
# Avg price per main_category x shipping
print("=== Avg Price: Category x Shipping ===")
pt1 = pd.pivot_table(df, values="price",
                    index="main_category",
                    columns="shipping",
                    aggfunc="mean",
                    fill_value=0).round(2)
pt1
```

```python
# Avg Price per main_category x condition_label
print("=== Avg Price: Category x Condition ===")
pt2 = pd.pivot_table(df, values="price",
                    index="main_category",
                    columns="condition_label",
                    aggfunc="mean",
                    fill_value=0).round(2)
pt2
```

```python
# Item count per main_category x shipping
print("\n Item Count: Category x Shipping ===")
ct1 = pd.crosstab(df["main_category"], df["shipping"])
ct1
```

```python
# Shipping rate per category (what % of items have seller shipping)
print("\n=== Shipping Rate per Category ===")
ct2 = pd.crosstab(df["main_category"], df["shipping"], normalize="index").round(3)
ct2
```
